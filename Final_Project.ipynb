## 2. Background

The USDA National Household Food Acquisition and Purchase Survey (FoodAPS) dataset has been selected for this analysis due to its relevance regarding dietary health in the United States. This comprehensive survey includes over 4,800 households across the US and contains information on food acquisition, dietary quality, food insecurity, health outcomes, food spending, demographics, and socioeconomic status. As such, this dataset can be leveraged to compare different survey categories and examine correlations surrounding the impacts of food access.

To ensure respondent confidentiality, individual data will not be disclosed. Moreover, caution will be exercised when working with demographic data to avoid reinforcing biases regarding food security in underserved communities. Finally, due to the limitations of survey results, the results will focus on correlation rather than causation, while understanding the complexities of food security.

The FoodAPS dataset is publicly accessible at the following link and is released under the Creative Commons Attribution 4.0 International License: [FoodAPS Dataset](https://www.ers.usda.gov/data-products/foodaps-national-household-food-acquisition-and-purchase-survey/foodaps-national-household-food-acquisition-and-purchase-survey/#Public-Use%20Data%20Files%20and%20Codebooks).  The data can be downloaded as a CSV file along with relevant documentation.

Previous research, such as studies by Cook and Frank [1] and Seligman et al. [2], have contributed insights into the challenges faced by underserved communities in accessing healthy and affordable food. Cook and Frank found that food insecurity led to unhealthier food choices, with a preference for processed and high-fat foods over fruits and vegetables [1]. Seligman et al. found that food insecurity was associated with diseases like hypertension and diabetes, due to the high accessibility of high-sugar and high-sodium foods [2]. Building upon this existing research through the FoodAPS dataset, I seek to enhance my understanding of the relationships between food access, purchasing habits, and dietary outcomes, providing valuable insights for addressing the obesity epidemic and improving food equity in the country.

## 4. Methodology

For this study, we will utilize two specific files from the dataset: 'faps_household_puf.csv' and 'faps_access_puf.csv'. These files contain relevant variables such as household characteristics and food acquisition details, and we import them using Python's pandas library, setting low_memory to False due to their size.

```python
import pandas as pd

# Import household and access datasets
household_df = pd.read_csv('faps_household_puf.csv', low_memory=False)
access_df = pd.read_csv('faps_access_puf.csv', low_memory=False)
```

In these, we will analyze the following demographic variables: household income, household size, presence of children, distance to SNAP retailers, distance to supercenters, and distance to fast-food outlets. These were chosen based on their relevance to the research questions related to food insecurity and dietary quality.

To begin analysis, we will examine the distribution of distances from households to SNAP retailers, helping us understand the proximity of households to these important food access points. To do this, we utilize matplotlib to generate a chart with labels. Next, we explore the distances from households to supercenters--large retail stores that offer a wide range of food and non-food products. Finally, we investigate the distances from households to fast-food outlets, stack all 3 charts on top of each other, and display them.

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Analyze household distances to SNAP retailers
snap_dist = access_df['nearsnap_dist']
axs[0].hist(snap_dist, bins=10, range=(0, 5), color='blue')
axs[0].set_xlabel('Distance to SNAP Retailers (miles)')
axs[0].set_ylabel('# of Respondents')
axs[0].set_title('Distribution of Household Distances to SNAP Retailers')

# Analyze household distances to supercenters
sm_dist = access_df['nearsmss_dist']
axs[1].hist(sm_dist, bins=10, range=(0, 5), color='green')
axs[1].set_xlabel('Distance to Supercenters (miles)')
axs[1].set_ylabel('# of Respondents')
axs[1].set_title('Distribution of Household Distances to Supercenters')

# Analyze household distances to fast food
ff_dist = access_df['nearff_dist']
axs[2].hist(ff_dist, bins=10, range=(0, 5), color='orange')
axs[2].set_xlabel('Distance to Fast Food Outlets (miles)')
axs[2].set_ylabel('# of Respondents')
axs[2].set_title('Distribution of Household Distances to Fast Food Outlets')

# Adjust spacing
plt.tight_layout()

# Display charts
plt.show()
```

Next, we'll clean the household dataframe, removing any negative values to ensure that we're working with valid and relevant data. This lets us calculate the descriptive statistics for demographic variables and print them as a table.

```python
# Clean data
household_df = household_df[household_df['grocerylistfreq'] >= 0]
household_df = household_df[household_df['fruitsveg'] >= 0]

# Calculate descriptive statistics for demographic variables
demographic_stats = household_df[['resunitsize', 'hhsize', 'famsize']].describe().transpose()
print(demographic_stats)
```

We'll perform a similar analysis on the dietary quality variables, printing them as a table as well.

```python
# Calculate descriptive statistics for dietary quality variables
dietary_quality_stats = household_df[['fruitsveg', 'nutritioneduc']].describe().transpose()
print(dietary_quality_stats)
```

Finally, we import the statsmodels library to conduct regression analysis (using the Ordinary Least Squares (OLS) method), and print the model summary.

```python
import statsmodels.api as sm

# Prepare data for regression analysis
X = sm.add_constant(household_df['foodsecureq1'])
y = household_df['fruitsveg']

# Fit regression model and display results
model = sm.OLS(y, X).fit()
print(model.summary())
```

## 5. Findings

This analysis of household data reveals important insights regarding food security and fruit and vegetable consumption:

* An average household size of approximately 2.94 individuals, with a standard deviation of 1.72--over half of surveyed households were comprised of 1-3 individuals.
* Households reported consuming an average of 1.74 servings of fruits and vegetables, with a standard deviation of 0.46--suggesting room for improvement in meeting dietary guidelines.
* The majority of households (approximately 94%) reported a nutrition education score of 0, highlighting the need for formal nutrition education.
* The regression analysis indicated a coefficient of -0.0728 for the 'foodsecureq1' variable, implying a negative and statistically significant relationship between food security and fruit and vegetable consumption (p-value < 0.001).

## 6. Discussion, Limitations, and Implications


These results reveal the effects of food insecurity on dietary quality in the United States--a strong association between food insecurity and low fruit & vegetable consumption.

However, it is important to acknowledge the limitations of this study:
* The analysis relies on self-reported data, which can be influenced by recall bias and social desirability bias.
* Additionally, the cross-sectional nature of the data limits the ability to establish any causal relationships, and any results are purely correlational.

To address these limitations, future research could employ longitudinal designs to examine the effects of food insecurity over time. Overall, the strong association between food insecurity and poor dietary quality shows a need for policy development aimed at addressing food insecurity in the United States, mitigating challenges regarding access to nutritious and affordable food, and promoting better nutrition among vulnerable populations.
